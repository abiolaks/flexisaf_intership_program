# Building my Preprocessing Pipeline for both Numeric and categorical


```python
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.preprocessing import OneHotEncoder, StandardScaler, FunctionTransformer
from sklearn.impute import SimpleImputer
from sklearn.ensemble import IsolationForest
from sklearn.base import BaseEstimator, TransformerMixin
```

```python

# -------------------------------
# 1. Custom cleaning transformer
# -------------------------------
def custom_cleaning(X):
    X = X.copy()
    if "roof" in X.columns:
        X["roof"] = (
            X["roof"]
            .replace("composition", "Composition")
            .replace("asphalt", "Asphalt")
            .replace(["shake-shingle", "asphalt,shake-shingle"], "Shake Shingle")
        )

    if "exterior_walls" in X.columns:
        X["exterior_walls"] = (
            X["exterior_walls"]
            .replace("Rock, Stone", "Masonry")
            .replace(["Concrete", "Block"], "Concrete Block")
        )
    return X

# Wrap in transformer

custom_cleaning_transformer = FunctionTransformer(custom_cleaning)

# -------------------------------
# 2. Outlier removal transformer
# -------------------------------
class OutlierRemover(BaseEstimator, TransformerMixin):
    def __init__(self, contamination=0.05):
        self.contamination = contamination
        self.model = IsolationForest(contamination=self.contamination, random_state=42)
    
    def fit(self, X, y=None):
        self.model.fit(X)
        self.inliers_ = self.model.predict(X) == 1
        return self
    
    def transform(self, X):
        return X[self.inliers_]
    
    def transform_xy(self, X, y):
        """Helper method to filter X and y together"""
        X_filtered = X[self.inliers_]
        y_filtered = y[self.inliers_]
        return X_filtered, y_filtered

# -------------------------------
# 3. Skewed log transformer
# -------------------------------
class SkewedLogTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, skew_threshold=1.0):
        self.skew_threshold = skew_threshold
        self.skewed_features_ = []
        
    def fit(self, X, y=None):
        # Work with DataFrame for easier column names
        X_df = pd.DataFrame(X)
        skewness = X_df.skew(numeric_only=True)
        self.skewed_features_ = skewness[skewness > self.skew_threshold].index.tolist()
        return self
    
    def transform(self, X):
        X_df = pd.DataFrame(X).copy()
        for col in self.skewed_features_:
            X_df[col] = np.log1p(X_df[col].clip(lower=0))  # avoid negatives
        return X_df.values

# -------------------------------
# 4. Pipelines for categorical & numeric
# -------------------------------
categorical_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

numeric_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("log", SkewedLogTransformer(skew_threshold=1.0)),  # apply only to skewed
    ("scaler", StandardScaler())
])

# -------------------------------
# 5. Column transformer with selectors
# -------------------------------
preprocessor = ColumnTransformer([
    ("categorical", categorical_pipeline, make_column_selector(dtype_include=object)),
    ("numerical", numeric_pipeline, make_column_selector(dtype_include=np.number))
])

# -------------------------------
# 6. Full pipeline
# -------------------------------
full_pipeline = Pipeline([
    ("cleaning", custom_cleaning_transformer),
    ("preprocessing", preprocessor),
    # ("outliers", OutlierRemover(contamination=0.05))  # run separately before train/test
])
```

```python
# Full preprocessing
X_train = train_set.drop("tx_price", axis=1)
y_train = train_set["tx_price"].copy()
X_test = test_set.drop("tx_price", axis=1)
y_test = test_set["tx_price"].copy()
```

```python
# 1. Fit pipeline to transform raw data
X_train_proc = full_pipeline.fit_transform(X_train)
X_test_proc  = full_pipeline.transform(X_test)

# 2. Run outlier removal (numeric already)
outlier_remover = OutlierRemover(contamination=0.05)
outlier_remover.fit(X_train_proc)

X_train_proc, y_train = outlier_remover.transform_xy(X_train_proc, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor

models = {
    "RandomForest": RandomForestRegressor(random_state=42),
    "XGBoost": XGBRegressor(random_state=42, n_estimators=500, learning_rate=0.05),
    "LightGBM": LGBMRegressor(random_state=42, n_estimators=500, learning_rate=0.05)
}
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
results = {}

for name, model in models.items():
    model.fit(X_train_proc, y_train)
    preds = model.predict(X_test_proc)

    results[name] = {
        "RMSE": np.sqrt(mean_squared_error(y_test, preds)),
        "MAE": mean_absolute_error(y_test, preds),
        "R2": r2_score(y_test, preds)
    }

results_df = pd.DataFrame(results).T
print(results_df)
```
